```python
data['Pedestrian'].fillna(data['Pedestrian'].mean(),inplace=True)
```

```python
data['Bicycles'].fillna(data['Bicycles'].mean(),inplace=True)
```

```python
data['Two-wheelers'].fillna(data['Two-wheelers'].mean(),inplace=True)
```
